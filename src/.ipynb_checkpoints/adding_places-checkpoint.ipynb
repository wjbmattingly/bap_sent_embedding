```python
import pandas as pd
```

```python
df = pd.read_csv("../data/testimonies_complete.csv")
df
```

```python
# files = [file.replace("\\", "/") for file in df.file.unique()]
hearing_types = []
places = []
nums = []
for idx, row in df.iterrows():
    file = row.file
    b,n,m,hearing_type, place, num = file.split("/")
    hearing_type = hearing_type.replace("_", " ").title()
    place = place.replace("_", " ").title()
    num = num.replace(".json", "")
    hearing_types.append(hearing_type)
    places.append(place)
    nums.append(num)
```

```python
df["hearing_type"] = hearing_types
df["location"] = places
df["file_num"] = nums
```
