```python
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
df = pd.read_csv("../data/speaker_dialogue.csv")
df
```

```python
line_locations = []
for idx, row in df.iterrows():
    with open(row.file, "r") as f:
        data = json.load(f)
    line = 1
    segment_location = {}
    for i, segment in enumerate(data["testimony"]):
        line=line+1
        segment_location[i] = line
        line = line+segment[1].count("\n")
    line_locations.append(segment_location[row.file_index])
```

```python
df["saha_loc"] = line_locations
```

```python
df
```

```python
df.to_csv("../data/testimonies_complete.csv")
```
